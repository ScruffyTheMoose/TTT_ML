```python
from TicTacToe import GameTools as gt
import pandas as pd
import random
```

```python
# generating a list of DataFrames
# each table will contain a 300 randomly generated matches with varying levels of progression

dataframes = list()

for ct in range(1, 9, 2):
    data = list()
    for _ in range(300):
        player = random.randint(0, 1)
        data.append(gt.randomized_match(first_player=player, open_pos=ct))
    dataframes.append(pd.DataFrame(data))

dataframes
```

```python
dataframes[0].to_csv('1_space_cluster_samples.csv', index=False)
dataframes[0]
```
